```python
import pandas as pd
import requests
import io
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FormatStrFormatter

url = 'https://raw.githubusercontent.com/cockytrumpet/data-mining-group-4/main/datasets/dataset.csv'
download = requests.get(url).content

df = pd.read_csv(io.StringIO(download.decode('utf-8')))
```

```python
df['Heartdiseaseorattack'] = [bool(item) for item in df['HeartDiseaseorAttack']]
df['HighBP'] = [bool(item) for item in df['HighBP']]
df.groupby([df['HeartDiseaseorAttack'],df['HighBP']]).size().plot.bar()
plt.title('Heart Attack or Heart Disease vs High Blood Pressure')
plt.show()
```

```python
df['HeartDiseaseorAttack'] = [bool(item) for item in df['HeartDiseaseorAttack']]
df['HighChol'] = [bool(item) for item in df['HighChol']]
df.groupby([df['HeartDiseaseorAttack'],df['HighChol']]).size().plot.bar()
plt.title('Heart Attack or Heart Disease vs High Cholesterol')
plt.show()
```

```python
A = df['Age']
##table = df.groupby([A,df['HeartDiseaseorAttack']]).count().unstack()
A = pd.cut(A, bins=np.r_[0 : A.max() + 1 : 1])

table = pd.crosstab(A, df['HeartDiseaseorAttack'])\
          .reindex(A.cat.categories, fill_value=0)
table.plot(kind='bar')
plt.title('Heart Disease or Attack based on Age Groups')
```
